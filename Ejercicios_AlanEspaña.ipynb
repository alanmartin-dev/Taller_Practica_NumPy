#1.1 List comprehension

```python
# Básica
cubos = [x**3 for x in range(1, 6)]
print(cubos)
# Con filtro
impares = [x for x in range(1, 15) if x % 2 != 0]
print(impares)
# Con bucle tradicional equivalente
lista = []
for x in range(1, 15):
    if x % 2 != 0:
        lista.append(x)
print(lista)
# Anidada
tabla = [[i + j for j in range(1, 4)] for i in range(1, 4)]
print (tabla)
# Transformación
nombres = ['Ana', 'JUAN', 'Pedro']
print(nombres)
minus = [n.lower() for n in nombres]
print(minus)
```

#MÓDULO 4 - Operaciones Numéricas y Manipulación de Datos



##4.1 Operaciones

```python
a = np.array([2,4,6])
b = np.array([1,3,5])

print(a + b)
print(a * b)
print(a ** 2)

print(a + 10)
```

##4.2 Broadcasting


```python
v = np.array([1,2,3])
M = np.ones((2,3))

print(M + v)

col = np.array([[1],[2]])
fil = np.array([10,20,30])

print(col + fil)
```

##4.3 ufuncs


```python
v = np.array([-1, -4, 9])

print(np.abs(v))
print(np.sqrt(np.abs(v)))
print(np.square(v))

pos = np.array([1,2,3])
print(np.log(pos))
print(np.exp(pos))
```

##4.4 Reducciones


```python
M = np.array([[1,2,3],[4,5,6]])

print(M.sum())
print(M.mean())

print(M.sum(axis=0))
print(M.sum(axis=1))

print(M.max(axis=1))
```

##4.5 Concatenación


```python
A = np.array([[1,2]])
B = np.array([[3,4]])

print(np.concatenate([A,B], axis=0))
print(np.concatenate([A,B], axis=1))

v1 = np.array([1,2])
v2 = np.array([3,4])

print(np.vstack([v1,v2]))
print(np.hstack([v1,v2]))
```
