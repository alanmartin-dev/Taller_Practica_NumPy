##Ejercicio 2: Dado el diccionario datos = {'a':10, 'b':25, 'c':7, 'd':18}

```python
datos = {'a':10, 'b':25, 'c':7, 'd':18}
```

###a) Crea una lista con las claves cuyo valor sea mayor a 12.

```python
new_list = []
for clave, valor in datos.items():
  if valor > 12:
   new_list = new_list + [clave]
print(new_list)
```

###b) Crea un nuevo diccionario con esas claves y sus valores al cuadrado.

```python
nuevo_diccionario = {}
for clave, valor in datos.items():
  if valor > 12:
    nuevo_diccionario[clave] = valor ** 2
print(nuevo_diccionario)
```

##Ejercicio 3: Define una función 'resumen(lista)' que retorne un diccionario con: longitud, suma, promedio, mínimo y máximo de la lista.

```python
def resumen(lista):
  return {
     'logitud': len(lista),
      'Suma ': sum(lista),
      'Promedio ':sum(lista)/len(lista),
      'Minimo ': min(lista),
      'Maximo': max(lista)}

resumen = resumen ([1,4,25,53,6])
print(resumen)
```

##Ejercicio 4: Usa map() para aplicar la función potencia(x, 3) a la lista [1,2,3,4,5].

```python
def potencia(x,n):
  return x ** n

lista = [1,2,3,4,5]
resultado = list(map(lambda x: potencia (x,3),lista))

print (resultado)
```

##Ejercicio 1: Crea una matriz 5×4 de enteros aleatorios entre 1 y 100 (seed=7).

```python
np.random.seed(7)
matrix = np.random.randint(1, 101, size=(5, 4))
print(matrix)
```

###a) Calcula la media, máximo y mínimo de cada columna.

```python
print(np.mean(matrix, axis=0))
print(np.max(matrix, axis=0))
print(np.min(matrix, axis=0))
```

###b) Calcula la suma acumulada de la primera fila.

```python
print(np.cumsum(matrix[0]))
```

####c) Normaliza la matriz para que cada columna tenga valores entre 0 y 1.
###Fórmula: (x - min_col) / (max_col - min_col)

```python
min_col = np.min(matrix, axis = 0)
max_col = np.max(matrix, axis = 0)
normalized_matrix = (matrix - min_col)/(max_col - min_col)
print(normalized_matrix)
```

###Ejercicio 4: Crea dos matrices 3×4 y concaténalas vertical y horizontalmente. Luego divide el resultado vertical en 3 partes iguales con vsplit.

```python
matrix_1 = np.random.randint(1,101, size=(3, 4))
matrix_2 = np.random.randint(1,101, size=(3, 4))

print("Matriz vertical: \n ", np.concatenate((matrix_1, matrix_2), axis=0))
print("Matriz horizontal: \n ",np.concatenate((matrix_1, matrix_2), axis=1))

vsplit_fun = np.vsplit(np.concatenate((matrix_1, matrix_2), axis=0), 3)
print("Este es el resultado con vsplit: \n ", vsplit_fun)
```
